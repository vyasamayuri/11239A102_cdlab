```python
!pip install ply
import ply.lex as lex
import sys

# Workaround for AttributeError in interactive environments
# by providing a dummy __file__ attribute if it doesn't exist
if not hasattr(sys.modules['__main__'], '__file__'):
    sys.modules['__main__'].__file__ = 'lex_dummy_file.py'

class MyLexer:
    tokens = ['ADD', 'ID', 'NUMBER']

    t_ADD = r'ADD'
    t_ID = r'[a-zA-Z_][a-zA-Z0-9_]*'
    t_NUMBER = r'\d+'

    t_ignore = ' \t'

    def t_newline(self, t):
        r'\n+'
        t.lexer.lineno += len(t.value)

    def t_error(self, t):
        print(f"Illegal character '{t.value[0]}'")
        t.lexer.skip(1)

lexer_instance = MyLexer()
lexer = lex.lex(object=lexer_instance)

lexer.input("ADD item1 2")
for tok in lexer:
    print(tok)
```
