```python
def generate_quadruples(expr):
    temp_counter = 1
    quads = []
    operand_stack = []
    operator_stack = []

    # Define operator precedence
    precedence = {'+': 1, '*': 2}

    # Split expression into tokens. Assuming tokens are space-separated.
    tokens = expr.split()

    for token in tokens:
        if token.isalpha() or token.isdigit():
            # If it's an operand (variable or number), push to operand stack
            operand_stack.append(token)
        elif token in ['+', '*']:
            # If it's an operator
            # Pop operators from operator stack with higher or equal precedence
            # and generate quadruples
            while (operator_stack and operator_stack[-1] in precedence and
                   precedence[operator_stack[-1]] >= precedence[token]):
                op = operator_stack.pop()
                # Ensure there are enough operands
                if len(operand_stack) < 2:
                    raise ValueError(f"Not enough operands for operator {op} in expression: {expr}")
                op2 = operand_stack.pop()
                op1 = operand_stack.pop()

                temp_var = f"T{temp_counter}"
                temp_counter += 1
                quads.append((op, op1, op2, temp_var))
                operand_stack.append(temp_var) # Push temporary result back
            operator_stack.append(token) # Push current operator
        # Add handling for parentheses if needed (not in this example)

    # After processing all tokens, pop any remaining operators from the operator stack
    # and generate quadruples
    while operator_stack:
        op = operator_stack.pop()
        if len(operand_stack) < 2:
            raise ValueError(f"Not enough operands for remaining operator {op} in expression: {expr}")
        op2 = operand_stack.pop()
        op1 = operand_stack.pop()

        temp_var = f"T{temp_counter}"
        temp_counter += 1
        quads.append((op, op1, op2, temp_var))
        operand_stack.append(temp_var)

    return quads

print(generate_quadruples("burger + fries * 2"))
```
